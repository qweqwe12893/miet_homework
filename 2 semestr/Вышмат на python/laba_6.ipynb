```python
import matplotlib.pyplot as plt
import numpy as np

def draw_area_between(f, g, a, b):
    x = np.linspace(a, b, 100)
    plt.plot(x, f(x), label="f(x)", color='lime')
    plt.plot(x, g(x), label="g(x)", color='magenta')
    plt.fill_between(x, f(x), g(x), where=f(x)<=g(x), hatch='|')
    plt.grid()
    plt.legend()

f = lambda x: x**2
g = lambda x: np.sqrt(2*x - x**2)

a = 0
b = 1.4

draw_area_between(f, g, a, b)
```
