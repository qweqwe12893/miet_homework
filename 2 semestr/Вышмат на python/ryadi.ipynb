```python
import matplotlib.pyplot as plt

def draw_partial_sums(an, n, str_for_title):
    
    s = [an(1)] * n
    
    for i in range(2, n+1):
        s[i-1] = an(i) + s[i-2]

    n_for_plot = [i for i in range(1, n+1)]
    
    plt.plot(n_for_plot, s, color='fuchsia', marker='o')
    plt.grid(color='lime')
    
    ax = plt.gca() # get current axes
    ax.set_facecolor('#ffff00')
    
    plt.title(f'Частичные суммы ряда {str_for_title}')
    

draw_partial_sums(lambda x: 1/x, 20, '1/x')    
```

```python
# ряд вида q^a для нескольких значений 0<q<1, q=1, q>1


base = [0.5, 1, 2]


def partial_sum(an, n):
    s = an(1)
    for i in range(2, n+1):
        s += an(i)
    return round(s, 1)


def draw_partial_sums(an, n, str_for_title):
    
    s = [an(1)] * n
    
    for i in range(2, n+1):
        s[i-1] = an(i) + s[i-2]

    n_for_plot = [i for i in range(1, n+1)]
    
    plt.plot(n_for_plot, s, color='fuchsia', marker='o')
    plt.grid(color='lime')
    
    ax = plt.gca() # get current axes
    ax.set_facecolor('#ffff00')
    
    plt.title(f'Частичные суммы ряда {str_for_title}')

    
for q in base:

    if q < 1:
        print(f"Ряд {q}^alpha сходится. Сумма ряда:", partial_sum(lambda alpha: q**alpha, 20))
        draw_partial_sums(lambda alpha: q**alpha, 20, f"{q}^alpha")
        
    elif q >= 1:
        print(f"Ряд {q}^alpha расходится")
```

```python
# ряд вида 1/(n^alpha)

alpha = [0.5, 1, 2]


def draw_partial_sums(an, n, str_for_title):
    
    s = [an(1)] * n
    
    for i in range(2, n+1):
        s[i-1] = an(i) + s[i-2]

    n_for_plot = [i for i in range(1, n+1)]

    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    plt.plot(n_for_plot, s, color='fuchsia', marker='o')
    plt.grid(color='lime')
    
    ax = plt.gca() # get current axes
    ax.set_facecolor('#ffff00')
    
    plt.title(f'Частичные суммы ряда {str_for_title}')

    
for power in alpha:
    if power <= 1:
        print(f"Ряд 1/n^{power} расходится")
        draw_partial_sums(lambda n: 1/n**power, 35, f"1/n**{power}")
        
    elif power > 1:
        print(f"Ряд 1/n^{power} сходится")
        draw_partial_sums(lambda n: 1/n**power, 35, f"1/n**{power}")
```
