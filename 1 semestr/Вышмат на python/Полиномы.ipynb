```python
import numpy as np

print('            Задание №2            ')
pol = [2, 0, 0, -3, 0, 0, 7, 0, -2]
res = []


print(f"\nКорни полинома равны:")
for root in np.roots(pol):
    print(f"{root : .3f}")
    res.append(root)


res = np.poly(res)

print(f"\nКоэффициенты полинома равны:")
n = max([pol[i]/res[i] for i in range(len(pol))])

for root in res:
    print(f"{root*n : .1f}")
```

```python
import numpy as np

print('            Задание №4            ')
pol1 = [1, -1, 0, -3, 0, -2]
pol2 = [1, 0, 0, 0, 0, -3, 1]

print('\nПроизводная от первого:')

i = len(np.polyder(pol1))-1
for node in np.polyder(pol1):
    if node == 0:
        pass
    elif i == len(np.polyder(pol1))-1:
        print(f'{round(node)}x^{i}', end='')
    elif i != 0:
        print(f' + {round(node)}x^{i}', end='')
    else:
        print(f'{round(node)}')
    i -= 1


print('\nПроизводная от второго:')

i = len(np.polyder(pol2))-1
for node in np.polyder(pol2):
    if node == 0:
        pass
    elif i == len(np.polyder(pol2))-1:
        print(f'{round(node)}x^{i}', end='')
    elif i != 0:
        print(f' + {round(node)}x^{i}', end='')
    else:
        print(f'{round(node)}')
    i -= 1

print('\nПроизводная от произведения:')

i = len(np.polyder(np.convolve(pol1, pol2)))-1
for node in np.polyder(np.convolve(pol1, pol2)):
    if node == 0:
        pass
    elif i == len(np.polyder(np.convolve(pol1, pol2)))-1:
        print(f'{round(node)}x^{i}', end='')
    elif i != 0:
        print(f' + {round(node)}x^{i}', end='')
    else:
        print(f'{round(node)}')
    i -= 1

print('\nПроизводная от частного:')

i = len(np.polydiv(pol1, pol2)[0])-1

for node in np.polyder(np.polydiv(pol1, pol2)[0]):
    if node == 0:
        pass
    elif i == len(np.polydiv(pol1, pol2)[0])-1:
        print(f'{round(node)}x^{i}', end='')
    elif i != 0:
        print(f' + {round(node)}x^{i}', end='')
    else:
        print(f'{round(node)}')
    i -= 1

i = len(np.polydiv(pol1, pol2)[1])-1

for node in np.polyder(np.polydiv(pol1, pol2)[1]):

    if node == 0:
        pass
    elif i == len(np.polydiv(pol1, pol2)[1])-1:
        print(f'{round(node)}x^{i}', end='')
    elif i != 0:
        print(f' + {round(node)}x^{i}', end='')

    else:
        print(f'{round(node)}')
    i -= 1
```
