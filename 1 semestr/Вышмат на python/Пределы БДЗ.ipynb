```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt


A = np.array([2, -3, 1])
B = np.array([0, -11, 3])
C = np.array([4, 5, -1])

AB = B - A
AC = C - A

if all(i == 0 for i in np.cross(AB, AC)):
    L = Line(A, C)
    print("Точки A, B, C лежат на одной прямой.")
    print(f"Уравнение это прямой: {L.equation()[0]}. \n")


    ax = plt.axes(projection ='3d')
    plt.title('Задание № 1')
    x = np.array([2, 0, 4])
    y = np.array([-3, -11, 5])
    z = np.array([1, 3, -1])
    
    ax.plot(x, y, z, color='cyan')
    ax.plot(*A, 'or')
    ax.plot(*B, 'or')
    ax.plot(*C, 'or')
    
    ax.text(*A+[0, 0, 0.5], 'A', color='red')
    ax.text(*B+[0, 0, 0.5], 'B', color='red')
    ax.text(*C+[0, 0, 0.5], 'C', color='red')
    plt.show()
    
```

```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
from matplotlib import cm

M = np.array([1, -7, 2])
norm  = np.array([1, -1, -3])

P = Plane(M, normal_vector=norm)
print(f"Уравнение плоскости: {P.equation()}. \n")


ax = plt.axes(projection ='3d')

A = 1
B = -1
C = -3
D = -2
f = lambda x,y:(-A*x-B*y-D)/C
xval = np.linspace(-8, 8, 100)
yval = np.linspace(-8, 8, 100)
x, y = np.meshgrid(xval, yval)
z = f(x,y)
surf = ax.plot_surface(x, y, z, rstride=10, cstride=10, cmap=cm.rainbow).set_alpha(0.5)

ax.plot(*M,'ro')
ax.quiver(*M, *norm, color='blue')
ax.text(*M+[0.5, 0.5, 0.5], 'M(1, -7, 2)')
ax.text(*[2+0.5, -8+0.5, -1-0.3], 'M(1, -7, 2)')
plt.title("Задание № 2")

ax.view_init(elev=40)

plt.show()
```

```python
import numpy as np
from sympy import *
import math
import matplotlib.pyplot as plt


n1 = np.array((2, 3, -4))
n2 = np.array((5, -2, 1))
cos = (np.dot(n1, n2) / np.linalg.norm(n1) / np.linalg.norm(n2))
cos = math.acos(cos)
cos = math.degrees(cos)
print(f"Угол между плоскостями равен {cos}° \n")


ax = plt.axes(projection='3d')
A1 = 2
B1 = 3
C1 = -4
D1 = 4
f1 = lambda x, y: (A1*x + B1*y + D1)/-1*C1

A2 = 5
B2 = -2
C2 = 1
D2 = -3
f2 = lambda x, y: (A2*x + B2*y + D2)/-1*C2

xval = np.linspace(-10, 10, 100)
yval = np.linspace(-10, 10, 100)
x, y = np.meshgrid(xval, yval)
z1 = f1(x, y)
z2 = f2(x, y)

surf = ax.plot_surface(x, y, z1, rstride=10, cstride=10, color='red').set_alpha(0.5)
surf = ax.plot_surface(x, y, z2, rstride=10, cstride=10, color='cyan').set_alpha(0.5)

ax.quiver(*(0, 0, 1), *n1)
ax.text(*n1+[1, 0, 0], 'n1 (2, 3, -4)')
ax.quiver(*(0, 0, 3), *n2)
ax.text(*n2+[1, 0, 0], 'n2 (5, -2, 1)')

plt.title("Задание № 3")
plt.show()
```

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

#  x - y + 4z - 6 = 0  при z = 0: (x - y = 6) & (2x + y = -3) Решение: (1, -5, 0)
# 2x + y -  z + 3 = 0  при x = 0: (-y + 4z = 6) & (y - z = -3) Решение: (0, -2, 1)
# направляющий вектор: (1, -3, -1)
# (x - 1)/1 = (y+5)/-3 = (z - 0)/-1

# Параметрическое уравнение
# t = x - 1,       x = t + 1
# t = (y + 5)/-3,  y = -3t - 5
# t = -z,          z = -t  

# x - 1 = (y + 5)/-3    -3x + 3 = y + 5    3x + y + 2 = 0    
# x - 1 = -z            x + z - 1 = 0

t = np.linspace(-10, 10, 10)
xline = t + 1
yline = -3*t - 5
zline = -1*t

ax = plt.axes(projection='3d')
A = 3
B = -1
C = 6
D = -12
f = lambda x, y: (-A*x - B*y - D)/ C
xval = np.linspace(-20, 20, 100)
yval = np.linspace(-20, 20, 100)
x, y = np.meshgrid(xval, yval)
z = f(x, y)
surf = ax.plot_surface(x, y, z, rstride=10, cstride=10, color='lime').set_alpha(0.5)
ax.plot(xline, yline, zline)
print("Прямая и плоскость параллельны")
plt.show()
```
