```python
import numpy as np
import matplotlib.pyplot as plt

ugol = np.linspace(0, 2*np.pi, 100)
z = np.linspace(-2, 2, 100)

Ugol, Z = np.meshgrid(ugol, z)

a = 1
b = 1
c = 1

X = np.sinh(Z) * np.cos(Ugol)
Y = np.sinh(Z) * np.sin(Ugol)

Z1 = np.sqrt(X**2 + Y**2 + c**2)
Z2 = -np.sqrt(X**2 + Y**2 + c**2)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z1, cmap='winter', alpha=0.5)
ax.plot_surface(X, Y, Z2, cmap='winter', alpha=0.5)

ax.set_xlabel('ось X')
ax.set_ylabel('ось Y')
ax.set_zlabel('ось Z')
ax.set_title('Двуполосный гиперболоид')
ax.set_box_aspect([1,1,1])  # одинаковые оси

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')

a = 11  
b = 10  

phi = np.linspace(0, 2 * np.pi, 200) 
theta = np.linspace(0, 2 * np.pi, 200) 

phi, theta = np.meshgrid(phi, theta)

x = (a + b * np.cos(phi)) * np.cos(theta)
y = (a + b * np.cos(phi)) * np.sin(theta)
z = b * np.sin(phi)

ax.plot_surface(x, y, z, cmap='inferno', edgecolor='none')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Torus')

ax.set_xlabel('ось X')
ax.set_ylabel('ось Y')
ax.set_zlabel('ось Z')
ax.set_title('Тор')
ax.set_box_aspect([1,1,1])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,50)
z = np.linspace(0,3,50)
[x,z] = np.meshgrid(x,z)
y = np.sqrt(1- x**2)

fig = plt.figure(figsize=(7,7))
ax = plt.axes(projection='3d')

ax.plot_wireframe(x,y,z, color='lime')
ax.plot_wireframe(x,-y,z, color='lime')

ax.set_xlabel('ось X', color='red')
ax.set_ylabel('ось Y', color='red')


plt.title("Эллиптический цилиндр x² + y² = 1")
plt.show()
```
