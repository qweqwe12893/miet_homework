```python
import matplotlib.pyplot as plt 
import numpy as np
from math import degrees
A = np.array([1, 3, -1])
B = np.array([2, -1, 4])
C = np.array([5, 0, 3])

AB = B - A
AC = C - A

cos = np.dot(AB, AC) / (np.linalg.norm(AB) * np.linalg.norm(AC))
sin = (1 - cos**2)**0.5

V = 0.5 * np.linalg.norm(AB) * np.linalg.norm(AC) * sin

print(f"""Объем треугольника: {V}""")
```

```python
from matplotlib import pyplot as plt

from mpl_toolkits.mplot3d.art3d import Poly3DCollection 
import numpy as np

ax = plt.axes(projection='3d')

v = np.array([[1, 2, 5], [3, -1, 0], [-1, 4, 3]]) # [[x1, x2, x3]...] 
verts =  [list(zip(v[0], v[1], v[2]))]  # [[x1, y1, z1]...]

ax.scatter(*v, color='red') 
ax.add_collection(Poly3DCollection(verts, facecolors='cyan', edgecolors='blue', alpha=0.3))

ax.text(0.8, 3, -2, 'A', color='red')
ax.text(1.5, -1, 4, 'B', color='red') 
ax.text(5.3, 0, 3, 'C', color='red')
```

```python
import numpy as np

A = np.array([4, 5, 6])
B = np.array([2, 5, 4])
C = np.array([-2, 7, 8])
D = np.array([0, 2, 6])

AB = B - A
AC = C - A
AD = D - A

matrix = np.array([AB, AC, AD])
V = abs(np.linalg.det(matrix) / 6)

print(f"""Объем пирамиды: {V}""")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection 

A = np.array([4, 5, 6])
B = np.array([2, 5, 4])
C = np.array([-2, 7, 8])
D = np.array([0, 2, 6])
verts = np.array([(A, B, C), (B, C, D), (A, B, D), (A, C, D)])


ax = plt.axes(projection='3d')
ax.scatter(*list(zip(A, B, C, D)), color='red')

ax.add_collection3d(Poly3DCollection(verts, 
 facecolors='cyan', linewidths=1, edgecolors='blue', alpha=0.25))

ax.set_zlim([4, 9])
ax.text(4.5, 5, 6, 'A', color='red')
ax.text(2.7, 4, 4, 'B', color='red')
ax.text(-2.2, 7, 8.5, 'C', color='red')
ax.text(-0.3, 1.2, 6, 'D', color='red')
```
