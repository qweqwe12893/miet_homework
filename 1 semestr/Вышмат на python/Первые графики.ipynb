```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)
y = np.exp(x)
plt.plot(x, y)

plt.xlabel('Возраст')
plt.ylabel('Социальная ответственность')
plt.grid()
plt.title('ПОДАРОК ДЛЯ ПУТИНА')  # $y=e^x$
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2.0* np.pi, 2.0 * np.pi, 100)
h1, = plt.plot(x, np.cos(x), label='y=cos(x)')
h2, = plt.plot(x, np.cos(x * 2.0),  label='y=cos(2x)')
h3, = plt.plot(x, np.cos(x / 2), label='y=cos(0.5x)')

plt.legend(handles=[h1, h2, h3], loc='upper right')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return abs(abs(x)-2)

plt.style.use('bmh')
x = np.linspace(-10.0, 10.0, 100)
h1, = plt.plot(x, f(x-2), label='y=x-2')
h2, = plt.plot(x, f(x+2), label='y=x+2')
h3, = plt.plot(x, f(x*2), label='y=2x')
h4, = plt.plot(x, f(x/2), label='y=0.5x')
h5, = plt.plot(x, f(-x), label='y=-x')

plt.legend(handles=[h1, h2, h3, h4, h5], loc='upper right')
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('bmh')

def f(x):
    return abs(abs(x)-2)

x = np.linspace(-10.0, 10.0, 100)

plt.figure(figsize=(10, 10))

plt.subplot(3, 2, 1)
h0, = plt.plot(x, f(x), color='lime')
plt.ylim(0, 8)
plt.title('y = x')

plt.subplot(3, 2, 2)
h1, = plt.plot(x, f(x-2), color='lime')
plt.ylim(0, 8)
plt.title('y = x - 2')

plt.subplot(3, 2, 3)
h2, = plt.plot(x, f(x+2), color='lime')
plt.ylim(0, 8)
plt.title('y = x + 2')

plt.subplot(3, 2, 4)
h3, = plt.plot(x, f(x*2), color='lime')
plt.ylim(0, 8)
plt.title('y = 2x')

plt.subplot(3, 2, 5)
h4, = plt.plot(x, f(x/2), color='lime')
plt.ylim(0, 8)
plt.title('y = 0.5x')

plt.subplot(3, 2, 6)
h5, = plt.plot(x, f(-x), color='lime')
plt.ylim(0, 8)
plt.title('y = -x')

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1.1, 1.1, 50)
y = [i**3 for i in x]
plt.style.use('Solarize_Light2')

plt.text(0.4, 1, 'y = x³', color='brown', size=14)
plt.plot(x, y, 'red')

plt.text(-0.5, 0.1, 'y = x³', color='red', size=14)
plt.plot(y, x, 'brown')
```
