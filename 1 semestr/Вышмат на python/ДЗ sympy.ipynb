```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt


L  = Line((2, 3), (-4, -1))
oy = Line((0, 0), (0, 1))
C = L.intersection(oy)[0]
print(f"Прямая, проходящая через точки (2, 3) и (-4, -1), пересекает ось y в точке ({C.x}, {C.y})")


plt.grid()
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

x = np.linspace(-10, 10, 100)
h, = plt.plot(x, (4*x + 10) / 6, color = 'cyan', linewidth=2, label='4x - 6y + 10 = 0')
plt.plot(0, 5/3, 'og')

plt.title('Задание № 1')
plt.legend(handles = [h])
plt.text(0.7, 1.45, 'C (0, 5/3)')

plt.show()
```
