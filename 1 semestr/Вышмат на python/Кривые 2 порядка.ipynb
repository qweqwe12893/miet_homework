```python
# 9x^2 + 4y^2 - 18x + 16y - 11 = 0
# 9x^2 - 18x + 9 + 4y^2 + 16y + 16 -36 = 0
# 9(x - 1)^2 + 4(y + 2)^2 = 36 | :36
# (x - 1)^2/4 + (y + 2)^2/9 = 1
# это уравнение эллипса

import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2*np.pi, 100)

x = 2 * np.cos(t) + 1
y = 3 * np.sin(t) - 2
c = (9 - 4) ** 0.5

plt.grid()
plt.axis('equal')
plt.axhline()
plt.axvline()

plt.plot(x, y, 'lime')
plt.plot(1, -2, 'or')
plt.text(1+0.3, -2+0.3, 'O(1, -2)', color='red')

plt.plot((1, 1), (-2+c, -2-c), 'or')
plt.text(1-0.7, -2+c+0.3, 'F1(1, 0.24)', color='red')
plt.text(1-0.7, -2-c+0.3, 'F2(1, −4.24)', color='red')


plt.title('Задание № 1', color='purple')
plt.ylabel('Ocь Y', color='magenta')
plt.xlabel('Ocь X', color='magenta')
plt.show()
```
