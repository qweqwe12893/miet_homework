```python
# lim (sin(x) / x)
from sympy import *

x = Symbol('x')
A = limit(sin(x) / x, x, 0)

print(A)
```

```python
x = Symbol('x')
A = limit((2 ** x - 1) / (x**2 - 3*x),x, 3, '+')
B = limit((2 ** x - 1) / (x**2 - 3*x),x, 3, '-')
print(A, B)

x = np.linspace(-2, 6, 1000)
f = (2 ** x - 1) / (x ** 2 - 3*x)

plt.plot(x, f, '.')
plt.grid()
```

```python
# f(x)= ((x **2 ) - 1) / (2 + x)
x = np.linspace(-10, 10, 100)
f = ((x ** 2) - 1) / (2 + x)

plt.plot(x, f, '.')
```

```python
from sympy import *

n = Symbol('n')

lim = limit(  ((n+3)**3 - (n-4)**3) / (5*n**2+ 23*n + 2) , n, oo)

a = solve_univariate_inequality(Abs( ((n+3)**3 - (n-4)**3) / (5*n**2+ 23*n + 2) - lim) < 0.001, n, relational=False)
b = solve_univariate_inequality(n > 0, n, relational=False)
c = Intersection(a, b)
pprint(c)
```
