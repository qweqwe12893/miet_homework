```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt


A = np.array([2, -3, 1])
B = np.array([0, -11, 3])
C = np.array([4, 5, -1])

AB = B - A
AC = C - A

if all(i == 0 for i in np.cross(AB, AC)):
    L = Line(A, C)
    print("Точки A, B, C лежат на одной прямой.")
    print(f"Уравнение это прямой: {L.equation()[0]}. \n")


    ax = plt.axes(projection ='3d')
    plt.title('Задание № 1')
    x = np.array([2, 0, 4])
    y = np.array([-3, -11, 5])
    z = np.array([1, 3, -1])
    
    ax.plot(x, y, z, color='cyan')
    ax.plot(*A, 'or')
    ax.plot(*B, 'or')
    ax.plot(*C, 'or')
    
    ax.text(*A+[0, 0, 0.5], 'A', color='red')
    ax.text(*B+[0, 0, 0.5], 'B', color='red')
    ax.text(*C+[0, 0, 0.5], 'C', color='red')
    plt.show()
    
```

```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
from matplotlib import cm

point = np.array([1, -7, 2])
norm  = np.array([1, -1, -3])

P = Plane(point, normal_vector=norm)
print(f"Уравнение плоскости: {P.equation()}. \n")


fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection='3d')

A = 1
B = -1
C = -3
D = -2
f = lambda x,y:(-A*x-B*y-D)/C
xval = np.linspace(-4, 4, 100)
yval = np.linspace(-4, 4, 100)
x, y = np.meshgrid(xval, yval)
z = f(x,y)
surf = ax.plot_surface(x, y, z, rstride=10, cstride=10, cmap=cm.rainbow).set_alpha(0.5)

plt.plot(*point,'ro')
ax.quiver(*point, *norm, color='blue')
ax.text(*point+[0.5, 0.5, 0.5], 'M(1, -7, 2)')
ax.text(*[2+0.5, -8+0.5, -1-0.3], 'M(1, -7, 2)')
plt.title("Задание № 2")

plt.show()
```

```python
import numpy as np
from sympy import *
import math
import matplotlib.pyplot as plt


n1 = np.array((2, 3, -4))
n2 = np.array((5, -2, 1))
cos = (np.dot(n1, n2) / np.linalg.norm(n1) / np.linalg.norm(n2))
cos = math.acos(cos)
cos = math.degrees(cos)
print(f"Угол между плоскостями равен {cos}° \n")


ax = plt.axes(projection='3d')
A1 = 2
B1 = 3
C1 = -4
D1 = 4
f1 = lambda x, y: (A1*x + B1*y + D1)/-1*C1

A2 = 5
B2 = -2
C2 = 1
D2 = -3
f2 = lambda x, y: (A2*x + B2*y + D2)/-1*C2

xval = np.linspace(-10, 10, 100)
yval = np.linspace(-10, 10, 100)
x, y = np.meshgrid(xval, yval)
z1 = f1(x, y)
z2 = f2(x, y)

surf = ax.plot_surface(x, y, z1, rstride=10, cstride=10, color='red').set_alpha(0.5)
surf = ax.plot_surface(x, y, z2, rstride=10, cstride=10, color='cyan').set_alpha(0.5)

ax.quiver(*(0, 0, 1), *n1)
ax.text(*n1+[1, 0, 0], 'n1 (2, 3, -4)')
ax.quiver(*(0, 0, 3), *n2)
ax.text(*n2+[1, 0, 0], 'n2 (5, -2, 1)')

plt.title("Задание № 3")
plt.show()
```

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

n1 = np.array([1, -1, 4])
n2 = np.array([2, 1, -1])
na = np.array([3, -1, 6])


def angle(x, y):
    cos = (np.dot(x, y) / np.linalg.norm(x) / np.linalg.norm(y))
    cos = math.acos(cos)
    cos = math.degrees(cos)
    return cos if cos <= 90 else 180-cos


print(f"Прямая x - y + 4z - 6 и плоскость пересекаются, угол между ними равен {round(angle(n1, na), 3)}")
print(f"Прямая 2x + y - z + 3 плоскость пересекаются, угол между ними равен {round(angle(n2, na), 2)}")

ax = plt.axes(projection='3d')
A = 3
B = -1
C = 6
D = -12
f = lambda x, y: (-A*x - B*y - D)/ C
xval = np.linspace(-10, 10, 100)
yval = np.linspace(-10, 10, 100)
x, y = np.meshgrid(xval, yval)
z = f(x, y)
surf = ax.plot_surface(x, y, z, rstride=10, cstride=10, color='lime').set_alpha(0.5)

# line1 (0, 0, 6/4) (0, -6, 0)
l1 = Line((0, 0, 6/4), (0, -6, 0))
plt.plot(xval, yval, (xval - yval -6)/-4)

# line2 (0, 0, 3) (0, -3, 0)
l2 = Line((0, 0, 3), (0, -3, 0))
print(l1.equation())
plt.plot(xval, yval, (-xval + yval + 3))
plt.show()
```
