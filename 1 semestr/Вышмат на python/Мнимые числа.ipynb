```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2+3j, 2-3j, -2-3j, -2+3j])
print(x)
plt.plot()
plt.scatter(np.real(x), np.imag(x), marker='*', color='red', s=100)
plt.axis('equal')
plt.axhline(y=0, color="cyan")
plt.axvline(x=0, color="cyan")
plt.grid()
plt.title("Первый шедевр для мамы")
plt.xlabel("Дейвствительная часть")
plt.ylabel("Мнимая часть")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2+3j, 2-3j, -2-3j, -2+3j])
orig = np.zeros(4)
plt.plot()
q = plt.quiver(orig, orig, np.real(x), np.imag(x), angles='xy', color='purple', scale_units='xy', scale=1)
plt.axis([-4, 4, -4, 4])
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np

z = 1 + 1j
r = abs(z)
phi = np.angle(z)
k = np.arange(1, 7)
zroot = r ** (1/6) *(np.cos((phi + 2.0* np.pi * k) /6)) + 1j * (np.sin((phi + 2.0* np.pi * k) /6))
print(zroot)
plt.plot()
plt.title("ШЕДЕВР ДЛЯ БАБУЛИ")
plt.plot(np.real(zroot), np.imag(zroot), '.', color='orange')
plt.axis('equal')
t = np.linspace(0, 2*np.pi, 100)
plt.plot(r**(1/6) * np.cos(t), r**(1/6) * np.sin(t))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rnd
x = 3 - 6 * rnd.rand(10000)
y = 3 - 6 * rnd.rand(10000)
z = x - 1j / y
L = (np.abs(z+ 1j) >= 1.5)
plt.plot(x[L], y[L], '.')
plt.axis('equal')
```
